```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Create a dataset with random data
x_data = np.linspace(0, 10, 100)
y_data = 2*x_data + 5 + np.random.randn(*x_data.shape)
```

```python
# Define the model
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(units=1, input_shape=[1])
])

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mse')

# Train the model
history = model.fit(x_data, y_data, epochs=1000)

# Make predictions with the trained model
y_pred = model.predict(x_data)
```

```python
# Plot the results
plt.scatter(x_data, y_data)
plt.plot(x_data, y_pred, color='r')
plt.show()
```
