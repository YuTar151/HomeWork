```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import torchviz
```

```python
# nX, nH1, nH2, nH3, nY = 5, 4, 3, 2, 1
nX, nH, nH1, nY = 2, 5, 3, 1
```

```python
model = nn.Sequential(
          nn.Linear(nX, nH),   
          nn.Sigmoid(),         
          nn.Linear(nH, nH1),   
          nn.Sigmoid(),         
          nn.Linear(nH1, nY),   
          nn.Sigmoid() )        
```

```python
class ThreeLayersNet(nn.Module):
    def __init__(self, nX, nH1, nH2, nY):        
        super(ThreeLayersNet, self).__init__()     # конструктор
         
        self.fc1 = nn.Linear(nX, nH1)             # створення параметрів першого шару
        self.fc2 = nn.Linear(nH1, nH2)            # створення параметрів другого шару
        self.fc3 = nn.Linear(nH2, nY)             # створення параметрів третього шару
          
    def forward(self, x):                         # задаємо прямий прохід
        x = self.fc1(x)                           # вихід першого шару
        x = nn.Sigmoid()(x)                       # сигмоїдна активація
        x = self.fc2(x)                           # вихід другого шару
        x = nn.Sigmoid()(x)                       # сигмоїдна активація
        x = self.fc3(x)                           # вихід третього шару
        x = nn.Sigmoid()(x)                       # сигмоїдна активація
        return x
          
model = ThreeLayersNet(2, 5, 3, 1)                 # створення екземпляру класу
```

```python
X = torch.rand (1200,2)                       
Y = (torch.sum((X - 0.5)**2, axis=1) < 0.1).float().view(-1,1)
```

```python
plt.figure (figsize=(5, 5))
plt.scatter(X.numpy()[:,0], X.numpy()[:,1], c=Y.numpy()[:,0],cmap=plt.cm.Paired)        
plt.show()
```
