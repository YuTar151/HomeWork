Завантажте зображення cat.jpg, що знаходиться в репозиторії лекції. Проведіть для цього зображення:
• декомпозицію по складових RGB-моделі
• декомпозицію по складових HSV-моделі
• декомпозицію по складових моделі CIE LAB

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as clr
```

```python
img = cv2.imread('cat.jpg')
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.axis('off')
plt.imshow(img_RGB)
plt.show()
```

```python
R = img_RGB.copy()
G = img_RGB.copy()
B = img_RGB.copy()
```

```python
R[:,:,(1,2)] = 0
plt.imshow(R)
plt.show()
```

```python
G[:,:,(0,2)] = 0
plt.imshow(G)
plt.show()
```

```python
B[:,:,(0,1)] = 0
plt.imshow(B)
plt.show()
```

```python
img_HSV = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(img_HSV)
plt.show()
```

```python
H = img_HSV.copy()
S = img_HSV.copy()
V = img_HSV.copy()
```

```python
H[:,:, (1,2)] = 255
H_RGB = cv2.cvtColor(H, cv2.COLOR_HSV2RGB)
plt.imshow(H)
plt.show()
```

```python
new_RGB = R + G + B
```

```python
plt.imshow(new_RGB)
plt.show()
```
